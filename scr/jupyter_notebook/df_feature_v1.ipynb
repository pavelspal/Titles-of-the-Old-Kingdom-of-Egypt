# Feature selection version 1

**Motivation:**  
This script creates **Feature version 1**.

**Version 1 definition:**
- Select all titles that appeared in the titulary of more than 10 viziers.  
    - That is, select titles satisfying: `count(title | vizier) > 10`.

### loads

```python
import numpy as np
import pandas as pd
import set_path
import supp.support_load as sl
from supp.support_load import read_csv
from supp.support_constants import VIZIER_IDS, PATH_DATA, PATH_DESCRIPTION
from supp.support_save import save_df
from supp.support_analyzer import make_excel_analysis
```

```python
# load database
dfs, dfs_name, dfs_export_date = sl.load_pickle()
iton, ntoi = sl.get_name_dicts(dfs_name)
dfs_export_date
```

```python
# load dataframe with binary encoded titles
vizier__35_title_list = read_csv('vizier__35_title_list')
vizier__35_title_list = vizier__35_title_list.set_index('ID_person', drop=True)
print(vizier__35_title_list.shape)
vizier__35_title_list.head()
```

```python
# read df sex of all persons
df_sex = read_csv('08_sex')
print(df_sex.shape)
df_sex.head()
```

### feature selection

```python
# RESTRICTION
# select only viziers
df1 = vizier__35_title_list[vizier__35_title_list['vizier'] == 1]
#df1 = df1.drop(columns=['vizier'])
print(f'{df1.shape}\tshape of df1')
df1.head(2)
```

```python
# show persons with the fewest titles
df_help = df1[['vizier']].copy()
df_help['titel_total_count'] = df1.sum(axis=1)
df_help = df_help.sort_values('titel_total_count', ascending=True)
df_help.head(20)
```

```python
# show persons with the most titles
df_help = df1[['vizier']].copy()
df_help['titel_total_count'] = df1.sum(axis=1)
df_help = df_help.sort_values('titel_total_count', ascending=False)
df_help.head(20)
```

**NOTE:**
- There is about 15 viziers with relatively small number of title (up to 10).
- On the other side, there are viziers with more like 50 titles (one with 102).

```python
# RESTRICTION
# take only titles that have accured more like 10 times
col_mask = (df1.sum(axis=0)>10)
df_vizier = df1.loc[:, df1.columns[col_mask]]
print(df_vizier.shape)
df_vizier.head()
```

### correlation analysis

```python
# Compute the correlation matrix
correlation_matrix = df_vizier.corr()
# Extract the upper triangle of the correlation matrix without the diagonal
mask = np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool)
upper_triangle = correlation_matrix.where(mask)
# Flatten and sort the correlations by absolute value
correlation_pairs = upper_triangle.unstack()   #.dropna()
sorted_correlations = correlation_pairs.abs().sort_values(ascending=False)
# Display top correlations
df_correlation = sorted_correlations.reset_index()
df_correlation.columns = ['title_1', 'title_2', 'correlation']

print("Top Correlation Pairs:")
df_correlation.head()
```

**NOTE:**
- The pair (`mdw rxyt`, `iwn knmwt`) has a correlation above 0.85.  
- For now, I will keep it in the dataset.  
- If it causes problems with multicollinearity, VIF will remove it.

### remove vizier titles

```python
vizier_ids = VIZIER_IDS
vizier_ids
```

```python
# get names of vizier titles
df_titles = dfs[ntoi['df_titles']]
vizier_title_name = df_titles.loc[df_titles['ID_title'].isin(vizier_ids), 'title'].to_list()
vizier_title_name
```

```python
columns_all = df_vizier.columns
non_veziers_titles = list(set(columns_all) - set(vizier_title_name))
df_features = df_vizier[non_veziers_titles].copy()
print(f'{df_vizier.shape}\tshape of df with all features')
print(f'{df_features.shape}\tshape of df with non-vizier features')
df_features.head()
```

```python
# Control that vizier titles were removed
# MUST BE FALSE
any(vizier_title in df_features.columns for vizier_title in vizier_title_name)
```

### make final dataframe

```python
# define list with most common titles among viziers 
top_vizier_titles = df_features.columns.to_list()
top_vizier_titles = list(set(top_vizier_titles) - set(['ID_person', 'vizier']))
print(f'{len(top_vizier_titles)}\ttitles selected as the final selection')
```

```python
# take only selected features into final dataframe
columns_to_keep = ['vizier'] + top_vizier_titles
df_result = vizier__35_title_list[columns_to_keep]
print(df_result.shape)
df_result.head()
```
